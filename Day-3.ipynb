```python
import matplotlib.pyplot as plt
import numpy as np
```

### Line plot

```python
x = np.array([3,11,16,23,27,34,48,55,56,67])
y = np.array([13,23,34,49,47,61,69,78,84,89])
```

```python
plt.plot(x,y)
```

```python
plt.title('My First plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))
plt.yticks(range(0,101,10))
plt.plot(x,y)
```

```python
z= np.array([87,77,75,71,54,33,21,23,19,10])
```

```python
plt.title('My Second plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))
plt.yticks(range(0,101,10))
plt.plot(x,z)
```

```python
plt.title('My Third plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.xlim(0,80)
plt.ylim(0,100)
plt.xticks(range(0,81,10))
plt.yticks(range(0,101,10))
# marker: . o  >  <      V  ^   D   *
plt.plot(x,y,marker='o', color='red' , ls=':' , lw=3, mfc='green' )
#marker face value line size , line width 
```

```python
plt.title('My Fourth plot')
plt.xlabel('First value')
plt.ylabel('Second value')
plt.grid()
plt.plot(x,y, label='2022' , color='blue')
plt.plot(x,z, label='2023' , color='green')
plt.legend() #indicates label given to data 
```

### Separate input and output variables

### Import the class

```python
from sklearn.linear_model import LinearRegression
```
